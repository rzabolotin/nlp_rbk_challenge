```python
list_data = []
for query in tqdm(data['link']):
    time.sleep(0.3)
    texts = ()
    try:
        URL = f"https://google.com/search?q={query}"
        headers = {"user-agent" : USER_AGENT}
        resp = requests.get(URL, headers=headers)
        if resp.status_code == 200:
            soup = BeautifulSoup(resp.content, "html.parser")
        if resp.status_code == 429:
            break
        URL_news_list = re.findall("https\:\/\/www.rbc.ru\/[0-9|A-z|\/]+", str(soup))
        for URL_news in URL_news_list:
            try:
                URL_news = URL_news[:8] + 'nsk.' + URL_news[12:]
                driver.get(URL_news)
                
                try:
                    news_elements = driver.find_element(By.CLASS_NAME, "article__header__info-block")
                    text_header = news_elements.text
                except:
                    text_header = ''
                try:
                    news_elements = driver.find_element(By.CLASS_NAME, "article__text__overview")
                    text_overview = news_elements.text
                except:
                    text_overview = ''
                    
                try:
                    news_elements = driver.find_element(By.CLASS_NAME, "article__main-image__title")
                    photo_text = news_elements.text
                except:
                    photo_text = ''
                    
                try:
                    full_text = [x.text for x in driver.find_elements(By.TAG_NAME, "p")]
                except:
                    full_text = ''
                    
                    
                
                if max(len(text_header), len(text_overview), len(full_text)) > 0:
                    texts = (URL_news, text_header, text_overview, photo_text, full_text) 
                break
            except:
                1
    except:
        1

    list_data  +=  [ texts ]
```
